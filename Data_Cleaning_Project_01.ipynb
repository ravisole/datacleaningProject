```
# Importing the libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```
df=df.drop(df.index[[0]])
df.head()
```

```
df=df.drop(df.columns[[10]], axis=1)
df.head()
```

```
df1=pd.read_excel("Lending Data.xlsx", sheet_name='1b')
```

```
df1=df1.drop(df1.index[[0]])
df1.head()
```

```
df1.drop(df1.columns[10], axis=1, inplace=True)
df1.head()
```

```
df2=pd.read_excel("Lending Data.xlsx", sheet_name='1c')
df2=df2.drop(df2.index[[0]])
df2.head()
```

```
df2.drop(df2.columns[[7,11]], axis=1, inplace=True)
df2.head()
```

```
# Before Data Cleaning
df=df.drop([94,95,96])
df1.drop([94,95,96], inplace=True)
df2.drop([94,95,96], inplace=True)
```

```
frames=[df, df1, df2]
beforeCleaning=pd.concat(frames)
beforeCleaning.shape
```

```
beforeCleaning.to_csv('Before Clening.csv')
```

```
df['Rank'].replace({'NR':np.NaN}, inplace=True)
df.tail()
```

```
#df['Rank'].interpolate(method='linear').tail(10)
df['Rank']=df.Rank.ffill()+df.groupby(df.Rank.notnull().cumsum()).cumcount()
```

```
df.tail()
```

```
df1['Rank'].replace({'NR':np.NaN}, inplace=True)
df1['Rank']=df1.Rank.ffill()+df1.groupby(df1.Rank.notnull().cumsum()).cumcount()
df1.head(10)
```

```
df1.tail()
```

```
df2['Rank'].replace({'NR':np.NaN}, inplace=True)
df2['Rank']=df2.Rank.ffill()+df2.groupby(df2.Rank.notnull().cumsum()).cumcount()
df2.head(10)
```

```
frames=[df, df1, df2]
df=pd.concat(frames)
df.shape
```

```
df['Lender Asset Size'].value_counts()
```

```
df['Lender Asset Size'].replace({'$10B-$50B           ':1, '>$10B               ':0, '>$50B               ':2}, inplace=True)
```

```
df.describe()
```

```
df.info()
```

```
df['CC Amount/TA1'].replace({'    .  ':np.NaN}, inplace=True)
df.tail(10)
```

```
df['CC Amount/TA1'].replace(np.NaN, df['CC Amount/TA1'].mean(), inplace=True)
```

```
df.tail(10)
```

```
df.replace('-', np.NaN, inplace=True)
```

```
df.replace(np.NaN, df.mean(), inplace=True)
```

```
df.tail()
```

```
df.to_csv('After Cleaning.csv', index=False)
```
